## **Pandas Series**
### **Querying the Series**
In pandas, you can access elements of a Series using two methods:

*   **Accessing Element from Series with Position**

In pandas, you can access elements from a Series using the index operator `[]`. By default, pandas assigns integer-based positional indices to the elements in a Series, starting from 0. So, you can use the integer index to access elements in the Series.


```
# Sample data for the Series
data = [10, 20, 30, 40, 50]
index = ['A', 'B', 'C', 'D', 'E']

# Creating a pandas Series
series = pd.Series(data, index=index)

# Accessing elements by position using index operator []
# The index positions are zero-based, so index 0 corresponds to the first element.
element_at_position_0 = series[0]  # Accessing the first element
element_at_position_2 = series[2]  # Accessing the third element
element_at_position_4 = series[4]  # Accessing the fifth element

print(element_at_position_0)  
print(element_at_position_2)  
print(element_at_position_4)  

```
Output:
```
10
30
50
```

*   **Accessing Element Using Label (index)**
To access an element from a Pandas Series using the index label (not position) with the index operator `[ ]`, you can pass the label inside the brackets.


```
# Sample data for the Series
data = [10, 20, 30, 40, 50]
index = ['A', 'B', 'C', 'D', 'E']

# Create the Pandas Series
series_data = pd.Series(data, index=index)

# Accessing element with index label 'C' using the index operator
result = series_data['C']

print(result)

```
Output:
```
30
```

## **Pandas Dataframe**
### **Querying the Dataframe using `query()` method**
The `query()` method in Pandas allows you to access rows and columns of a DataFrame using a query expression. It is a convenient way to filter and retrieve specific data from a DataFrame. The `query()` method takes a string as its argument, which represents the query expression.


```
# Sample data for the DataFrame

    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
    'Age': [25, 30, 22, 28, 35],

df = pd.DataFrame(data)

# Accessing rows using query()
filtered_rows = df.query('Age > 25')

print(filtered_rows)

# Accessing specific columns using query()
selected_columns = df.query('Age > 25')[['Name', 'City']]

print("\nSelected columns where Age > 25:")
print(selected_columns)

```

```

### **Megrging**

When dealing with very large DataFrames and the need to combine them, joins become a powerful and efficient method to perform such operations. Joins are carried out on two DataFrames, referred to as the left and right tables, and they are based on a common column, known as the key. Choosing a key with unique values throughout the column helps prevent unintended duplication of row values.

In Pandas, the `merge()` function is the primary tool for performing standard database join operations between DataFrame objects. It allows for four basic types of joins: inner join, left join, right join, and outer join, each determining which rows will retain their data from the original DataFrames.


1.   **Merging a dataframe with one unique key combination**

To merge two DataFrames with a unique key combination, you can use the `pd.merge()` function in Pandas. The merge operation combines rows from both DataFrames based on the matching values of the specified key(s).


```
# Sample data for the first DataFrame
data1 = {
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
}

# Sample data for the second DataFrame
data2 = {
    'ID': [2, 3, 4, 5],
    'Age': [25, 30, 22, 28],
}

# Create the first DataFrame
df1 = pd.DataFrame(data1)

# Create the second DataFrame
df2 = pd.DataFrame(data2)

# Merge the two DataFrames using a unique key combination (ID)
merged_df = pd.merge(df1, df2, on='ID')

print(merged_df)

```
Output:

```
   ID     Name  Age
0   2      Bob   25
1   3  Charlie   30
2   4    David   22

```
2.   **Merging dataframe using multiple join keys**

To merge DataFrames using multiple join keys, you can provide a list of column names as the on parameter in the `pd.merge()` function. The merge operation will then combine rows from both DataFrames based on the matching values of the specified key combination.


```
# Sample data for the first DataFrame
data1 = {
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'City': ['New York', 'San Francisco', 'London', 'Tokyo']
}

# Sample data for the second DataFrame
data2 = {
    'ID': [2, 3, 4, 5],
    'Age': [25, 30, 22, 28],

}

# Create the first DataFrame
df1 = pd.DataFrame(data1)

# Create the second DataFrame
df2 = pd.DataFrame(data2)

# Merge the two DataFrames using multiple join keys (ID and City)
merged_df = pd.merge(df1, df2, on=['ID', 'City'])

print(merged_df)

```
Output:

```
   ID  Name           City  Age
0   3  Charlie        London   30
1   4    David         Tokyo   22

```
3. **Merging dataframe using how in an argument**

When merging DataFrames in Pandas, the how parameter allows you to specify the type of join operation to use.

|Merge Method|Join Name|Description|
|-------------|---------|-----------|
|inner|Inner Join|returns the common rows between both DataFrames |
|left|Left Join|returns all the rows from the left DataFrame and the matching rows from the right DataFrame|
|right|Right Join|returns all the rows from the right DataFrame and the matching rows from the left DataFrame|
|outer|Outer Join|returns all the rows from both DataFrames |

### **Joining**

The `.join()` function in Pandas is used to merge two DataFrames based on their index or specified column(s). It provides a convenient way to combine DataFrames with different columns or overlapping columns, as it automatically aligns the data based on the shared index or specified columns.


```

# Sample data for the first DataFrame
data1 = {
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
}

# Sample data for the second DataFrame
data2 = {
    'ID': [2, 3, 4],
    'Age': [25, 30, 22],
}

# Create the first DataFrame
df1 = pd.DataFrame(data1, index=['A', 'B', 'C'])

# Create the second DataFrame
df2 = pd.DataFrame(data2, index=['B', 'C', 'D'])

# Join the two DataFrames based on the index
joined_df = df1.join(df2, lsuffix='_left', rsuffix='_right')

print(joined_df)
```

Output:
```
   ID_left     Name  ID_right   Age
A        1    Alice       NaN   NaN
B        2      Bob       2.0  25.0
C        3  Charlie       3.0  30.0
```
### **Concatenating**

Both `.concat()` and `.append()` are used to concatenate DataFrames in Pandas, but they have some differences in how they handle the concatenation process.

1. `pd.concat()`:
The `pd.concat()` function is a powerful method for concatenating multiple DataFrames either vertically (along rows) or horizontally (along columns). It allows you to concatenate two or more DataFrames together, even if they have different columns.

```

# Sample data for the first DataFrame
data1 = {
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
}

# Sample data for the second DataFrame
data2 = {
    'ID': [4, 5],
    'Name': ['David', 'Emily'],
    'Age': [25, 30]
}

# Create the DataFrames
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Concatenate vertically (along rows)
concatenated_df = pd.concat([df1, df2], ignore_index=True)

print(concatenated_df)
```

Output:
```
   ID     Name   Age
0   1    Alice   NaN
1   2      Bob   NaN
2   3  Charlie   NaN
3   4    David  25.0
4   5    Emily  30.0
```

2. `DataFrame.append()`:
The `DataFrame.append()` method is a more limited version of `.concat()`, specifically designed for appending rows to an existing DataFrame.


```

# Sample data for the first DataFrame
data1 = {
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
}

# Sample data for the second DataFrame
data2 = {
    'ID': [4, 5],
    'Name': ['David', 'Emily'],
    'Age': [25, 30]
}

# Create the first DataFrame
df1 = pd.DataFrame(data1)

# Create the second DataFrame
df2 = pd.DataFrame(data2)

# Append df2 to df1
appended_df = df1.append(df2, ignore_index=True)

print(appended_df)
```

Output:
```
   ID     Name   Age
0   1    Alice   NaN
1   2      Bob   NaN
2   3  Charlie   NaN

4   5    Emily  30.0
```

















##**Pandas Groupby Operation**



Grouping.webp

In Pandas, the `groupby()` operation is used to group data in a DataFrame based on one or more columns, and then perform aggregate functions on those groups. It allows you to split the data into groups based on specific criteria and then apply functions to each group.


*   **Grouping data with one key**

 We use the `groupby()` function with the column name as the argument. This will group the data based on the unique values in the specified column and allow you to perform aggregate operations on each group.

```
# Sample data for the DataFrame
data = {
    'City': ['New York', 'New York', 'San Francisco', 'San Francisco', 'Chicago', 'Chicago'],
    'Temperature': [70, 75, 62, 68, 80, 85],
    'Humidity': [50, 45, 60, 55, 40, 35]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Grouping the data by the 'City' column
grouped_data = df.groupby('City')

# Finding the mean, minimum, and maximum values for each group
mean_values = grouped_data.mean()
min_values = grouped_data.min()
max_values = grouped_data.max()

print("Mean values:")
print(mean_values)

print("\nMinimum values:")
print(min_values)

print("\nMaximum values:")
print(max_values)

```

Maximum values:
               Temperature  Humidity
City

*   **Grouping data with multiple keys**

To group data with multiple keys in the above DataFrame and find the mean for each group, you can pass a list of column names to the `groupby()` function. This will group the data based on the unique combinations of values in the specified columns.

*   **Grouping data by sorting keys**

When you set the `sort` parameter to `True` while grouping data using the `groupby()` function, the groups will be sorted in ascending order based on the grouping columns. By default, `sort` is set to `True`, so the groups will be sorted unless you explicitly set it to `False`.


```
# Sample data for the DataFrame
data = {
    'City': ['New York', 'New York', 'San Francisco', 'San Francisco', 'Chicago', 'Chicago'],
    'Month': ['January', 'February', 'January', 'February', 'January', 'February'],

    'Humidity': [50, 45, 60, 55, 40, 35]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Grouping the data by multiple keys: 'City' and 'Month' with sorting enabled
grouped_data = df.groupby(['City', 'Month'], sort=True)

# Finding the minimum temperature and humidity for each group
min_values = grouped_data.min()

print(min_values)

```

Output:

```
                     Temperature  Humidity

             January           80        40
New York     February          75        45
             January           70        50
San Francisco February          68        55
             January           62        60

```
*   **Groupby and multiple aggregations**

It allows to group data in a DataFrame based on one or more columns and then apply multiple aggregation functions to each group simultaneously. This enables you to obtain a comprehensive summary of data for each group.

```
# Sample data for the DataFrame
data = {
    'Category': ['A', 'A', 'B', 'B', 'A', 'B', 'A', 'B'],
    'Value': [10, 20, 15, 25, 5, 30, 12, 18]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Grouping the data by the 'Category' column and applying multiple aggregations
grouped_data = df.groupby('Category').agg({
    'Value': ['mean', 'min', 'max', 'sum']
})

print(grouped_data)

```
Output:

```

##**Pandas Pivot table**
The versatile Pandas library provides a wide range of functions, including the powerful `pivot_table` function, which allows data scientists to create concise and informative two-dimensional tables summarizing the values of a feature. Similar to the `groupby()` function, the `pivot_table` function in Pandas is a valuable tool for data analysis.

Pivot tables play a crucial role in exploratory data analysis, providing flexibility and convenience to data scientists when they need to extract valuable insights from complex datasets. With pivot tables, data scientists can quickly aggregate, summarize, and visualize data, making it easier to understand patterns and relationships within the data.


The syntax for creating a pivot table in Pandas is as follows:

```
DataFrame.pivot_table(values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, margins_name='All')
```

Parameters:
- `values`: The column to aggregate. This column's values will be summarized in the pivot table.
- `index`: The column(s) used to create the rows of the pivot table.
- `columns`: The column(s) used to create the columns of the pivot table.
- `aggfunc`: The aggregation function to be applied to the `values` column. Default is `'mean'`.
- `fill_value`: The value to replace missing values in the pivot table.
- `margins`: If `True`, adds row/column totals to the pivot table.
- `margins_name`: The name of the row/column representing the totals when `margins=True`.



```
# Sample data for the DataFrame
data = {

5. **How to resample time-series data**

Resampling time-series data involves changing the frequency of data points, such as converting daily data to monthly or yearly data. Pandas provides the `resample()` method to facilitate this process.


```
# Sample time-series data
data = {
    'Date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    'Temperature': [25, 28, 30, 27, 26, 29, 31, 32, 29, 28],
}

# Create a DataFrame with DateTimeIndex
df = pd.DataFrame(data)
df.set_index('Date', inplace=True)

# Resample the data to monthly mean
monthly_mean = df.resample('M').mean()

print(monthly_mean)

5. **Handle Duplicates in Dataset**

As there are no duplicate data in the titanic dataset, add a duplicated row to the Titanic dataset and then handle the duplicates.

we add a duplicated row by copying the first row of the DataFrame and appending it to the DataFrame using `df.append()`. we handle the duplicates using `df.drop_duplicates()`, which removes duplicate rows from the DataFrame. We set `inplace=True` to modify the DataFrame in place.

````python
# Step 1: Add a duplicated row
duplicated_row = df.iloc[0].copy()  # Copy the first row
df = df.append(duplicated_row, ignore_index=True)  # Append the duplicated row

# Display the number of rows before handling duplicates
print("Number of rows before handling duplicates:", len(df))

# Step 2: Handle duplicates
df.drop_duplicates(inplace=True)

# Display the number of rows after handling duplicates
print("Number of rows after handling duplicates:", len(df))
````
